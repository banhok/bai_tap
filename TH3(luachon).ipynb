```python
#W3A11 (kiểm tra loại tam giác)
a, b, c = map(int, input("Nhập vào 3 số nguyên dương: ").split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

```python
#W3A12 (năm nhuận)
n=int(input("Nhập vào 1 năm:"))
if (n%4==0 and n%100!=0) or (n%400==0):
    print("Yes")
else:
    print("No")
```

```python
#W3A14 (giải phương trình bậc nhất)
a,b=map(float,input("Nhập vào 2 số thực a và b:").split())
if a==0:
    if b==0:
        print("Vô số nghiệm")
    else:
        print("Vô nghiệm")
else:
    x = -b / a
    print(f"Nghiệm của phương trình là: {x:.2f}")
```

```python
#W3A15 (xếp loại học lực)
diem=float(input("Nhập vào điểm trung bình:"))
if diem<5:
    print("Yếu")
elif 5<=diem<6.5:
    print("Trung bình")
elif 6.5<=diem<8:
    print("Khá")
else:
    print("Giỏi")
```

```python
#W3A16 (làm tròn số thực)
n=float(input("Nhập vào số thực n:"))
x=int(n)
lam_tron_len=x+1
lam_tron_xuong=x
lam_tron_so_nguyen_gan_nhat=0

if n-x>=0.5:
    lam_tron_so_nguyen_gan_nhat=lam_tron_len
else:
    lam_tron_so_nguyen_gan_nhat=lam_tron_xuong
print(lam_tron_len, lam_tron_xuong, lam_tron_so_nguyen_gan_nhat)
```
