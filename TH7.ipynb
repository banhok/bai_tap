```python
#W7A3.1
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        swapped = False
        for j in range(n-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

arr=[64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(arr))
```

```python
#W7A3.2
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_idx=i
        for j in range (i+1, n):

            if arr[min_idx] > arr[j]:
                min_idx=j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr

arr=[64, 34, 25, 12, 22, 11, 90]
print(selection_sort(arr))
```

```python

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i+=1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i+1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi+1, high)
    return arr

arr = [2,6,7,4,9]

print(quick_sort(arr, 0, len(arr)-1))
```

```python
#W7A1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid] < target:
            left = mid +1
        else:
            right = mid -  1 
    return -1

arr=list(map(int, input().split()))
target=int(input())
print(binary_search(arr, 0, len(arr) - 1, target))
```

```python
#W7A4
def max_frequence(arr):
    d={}
    for i in arr:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
    max_val=None
    max_fre=0
    for j in d:
        if d[j] > max_fre:
            max_fre = d[j]
            max_val = j

    return max_val, max_fre

arr=list(map(int, input().split()))
k,h=max_frequence(arr)

print(f"{k} xuat hien nhieu nhat, som nhat, {h}")
```

```python
#W7A5
def count(arr, x):
    dem=0
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] + arr[j] == x:
                dem+=1

    return dem

arr=list(map(int, input().split()))
x=int(input())
print(count(arr, x))
```

```python
#W7A7 
import ast
def lis_length(arr):
    n = len(arr)
    dp = [1] * n 

    for i in range(n):
        for j in range(i):
            if arr[j] < arr[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)

arr=ast.literal_eval(input())
print(arr)
print(lis_length(arr))


```

```python
#W7A8
def binary_search_app(arr, left, right, target):
    while left <= right:
        mid = (left+right)//2
        if arr[mid]==target:
            return len(arr)
        elif arr[mid] < target:
            left = mid +1
        else:
            right = mid -  1 
    return -1

A=[[2, 8], [1, 5], [10, 15]] 
B=[1, 4, 6, 11, 20]

for i in B:
    min_len = float('inf')
    for j in A:
        C=list(range(j[0], j[1]+1))
        ans=binary_search_app(C, 0, len(C)-1, i)
        if ans != -1:
            min_len = min(min_len, ans)

    print(min_len if min_len != float('inf') else -1, end = ' ')
```

```python
#W7A10
import bisect

def tong_mang_con_lon_nhat_mod_m(nums, m):
    prefix = 0
    ds_prefix = []
    ket_qua = 0

    for x in nums:
        prefix = (prefix + x) % m

        ket_qua = max(ket_qua, prefix)

        i = bisect.bisect_right(ds_prefix, prefix)
        if i < len(ds_prefix):
            ket_qua = max(ket_qua, (prefix - ds_prefix[i] + m) % m)

        bisect.insort(ds_prefix, prefix)

    return ket_qua

nums=[7, 11, 14, 2, 8, 4]
m=12
print(tong_mang_con_lon_nhat_mod_m(nums, m))
```
