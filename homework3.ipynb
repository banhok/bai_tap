```python
#bai4
h=abs(int(input()))
for i in range(h+1):
    l=h-i-1
    r=h+i-1
    for j in range(2*h-1):
        if j==l or j==r or i==h-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
#bai6
import math
a=int(input("Nhập vào số nguyên dương a:"))

for i in range(1, math.isqrt(a)+1):
    if a%i==0:
        print(i, end=" ")
        if i!=a//i:
            print(a//i, end=" ")
```

```python
#bai14

solonnhat = None
sonhonhat = None

while True:
    x = int(input("Nhập số nguyên (nhập -1 để dừng): "))
    if x == -1:
        break
    if solonnhat is None or x > solonnhat:
        solonnhat = x
    if sonhonhat is None or x < sonhonhat:
        sonhonhat = x

if solonnhat is None:
    print("Không có số nào được nhập!")
else:
    print("Số lớn nhất là:", solonnhat)
    print("Số nhỏ nhất là:", sonhonhat)
```
