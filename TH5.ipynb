```python
#bai 1
def maxx(a, b):
    if a>b:
        return a
    else:
        return b
    
a, b = map(int, input().split())
print(maxx(a, b))
```

```python
#bai 2
def swap(a, b):
    a, b = b, a
    return a, b
a, b = map(int, input().split())
tup=tuple(swap(a, b))
print(tup)
```

```python
#bai 3
import math
def snt(n):
    if n<2:
        return False
    else:
        for i in range(2, int(math.sqrt(n))+1):
            if n%i==0:
                return False
        return True
            
n=int(input())
print(snt(n))
            
```

```python
#bai 4
import math
def shh(n):
    tong=0
    for i in range(1, n):
        if n%i==0:
            tong+=i
    if n==tong:
        return True
    else:
        return False
    
n=int(input())
print(shh(n))


```

```python
#bai 5
def bai5(A, k):
    for i in range(len(A)):
        if A[i]==k:
            return i+1
    return -1
A = list(map(int, input().split()))
k = int(input())
print(A)
print(k)
print(bai5(A, k))
```

```python
#bai 6
def giaith(n):
    tich=1
    for i in range(1, n+1):
        tich*=i
    return tich

n=int(input())
print(giaith(n))
```

```python
#bai 7
def tinh(a,b,o):
    if o == '+':
        return a+b
    if o == '-':
        return a-b
    if o == '*':
        return a*b
    if o == '/':
        return a/b

A=list(input().split())
a=float(A[0])
b=float(A[1])
o=A[2]
print(tinh(a,b,o))
```

```python
#bai 8
def hamming(x, y):
    a=x^y
    dem=0
    while a>0:
        if a&1:
            dem+=1
        a>>=1
    return dem
x,y=map(int, input().split())
print(hamming(x, y))

```

```python
#bai 8 cach 2(Brian Kernighan)
def hamming(x, y):
    a=x^y
    dem=0
    while a>0:
        a = a&(a-1)
        dem+=1
    return dem
x,y=map(int, input().split())
print(hamming(x, y))
```

```python
#bai 9
def tongchuso(n):
    n=abs(n)
    tong=0
    while n!=0:
        tong=tong+(n%10)
        n=n//10
    return tong

n=int(input())
print(tongchuso(n))


```
