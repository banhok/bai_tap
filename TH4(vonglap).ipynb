```python
#W4A6 (số nguyên tố trong đoạn)
import math
a, b=map(int, input("Nhập vào hai số nguyên a,b (a<=b):").split())
tong=0  
for num in range(a, b+1):
    if num>1:
        snt=True
        for i in range(2, int(math.sqrt(num))+1):
            if num%i==0:
                snt=False
                break
        if snt:
            tong+=num
print("Tổng các số nguyên tố trong đoạn [{}, {}] là: {}".format(a, b, tong))
```

```python
#W4A8 (palindrome)
n=int(input("Nhập vào số nguyên dương n:"))
dem=0
while str(n) != str(n)[::-1]:
    reversed_n = int(str(n)[::-1])
    n+=reversed_n
    dem+=1
print("Số palindrome là:", n)
print("Số bước là:", dem)
```

```python
#W4A10 "[LongestCollatzUpToN] 
n=int(input("Nhập vào số nguyên dương n:"))
max_length=0
result_number=0
for i in range(1, n+1):
    length=1
    temp=i
    while temp!=1:
        if temp%2==0:
            temp=temp//2
        else:
            temp=3*temp+1
        length+=1
    if length>max_length:
        max_length=length
        result_number=i
print(result_number, max_length)
```

```python
#WA10 (collatz dùng hàm)

def length_collatz(x):
    dem=1
    while x!=1:        
        if x%2==0:
            x=x/2            
        else:
            x=x*3+1
        dem+=1
    return dem

def max_length_collatz(n):
    max_x=1
    max_len=1
    for x in range(1,n+1):
        L=length_collatz(x)
        if L>max_len:
            max_len=L
            max_x=x
        elif L==max_len and x<max_x:
            max_x=x
    return max_x, max_len
        


if __name__ == '__main__':
    n=int(input())
    print(max_length_collatz(n))
    
    
```
