```python
#W6A17
n=int(input("n hàng n cột"))
A=[]
for i in range(n):
    hang=list(map(int,input().split()))
    A.append(hang)

print("đường chéo chính: ", [A[i][i] for i in range(n)])
print("đường chéo phụ: ", [A[i][n-1-i] for i in range(n)])
```

```python
#W6A18
n, m, k = map(int, input("n hàng, m cột, k nguyên").split())
A=[]
for i in range(n):
    hang=list(map(int, input().split()))
    A.append(hang)
if 0<=k<m:
    tong_cot_k=sum(A[i][k] for i in range(n))
    print(tong_cot_k)
else:
    print(f"A không có cột {k}")
```

```python
#W6A23

def xoa_nho_hon_x(A,x):
    return [i for i in A if i>=x]

A=list(map(int, input().split()))
x=int(input())
print(xoa_nho_hon_x(A,x))
```

```python
#W6A24
def f(A):
    B=[]
    for i in A:
        if i != 0:
            B.append(i)
    dem=0
    for j in A:
        if j == 0:
            dem+=1
    for k in range(dem):
        B.insert(0, 0)
    return B
A=list(map(int, input().split()))
print(f(A))
```
