```python
#W6A4
A=input("Nhập cặp key:value cách nhau bằng dấu cách")
B=A.split()
C={}
for b in B:
    key, value = b.split(":")
    if key in C:
        C[key].append(value)

    else:
        C[key]=[value]

print(C)

```

```python
#W6A5
dayso=list(map(int, input().split()))
A={'zeros':0, 'positives':0, 'negatives':0}
for so in dayso:
    if so == 0:
        A['zeros']+=1
    if so >0:
        A['positives']+=1
    if so <0:
        A['negatives']+=1
print(A)



```
