```python
#W6A4
A=input("Nhập cặp key:value cách nhau bằng dấu cách")
B=A.split()
C={}
for b in B:
    key, value = b.split(":")
    if key in C:
        C[key].append(value)

    else:
        C[key]=[value]

print(C)
```

```python
#W6A9
def la_palindrome(s):
    s = s.lower().replace(' ', '')
    return s == s[::-1]

s=input()
print(la_palindrome(s))
```

```python
#W6A10
def cap(A,k):
    kqua=[]
    n=len(A)
    for i in range(n):
        for j in range(i+1, n):
            if A[i]+A[j]==k:
                kqua.append((A[i],A[j]))
    kqua.sort(key=lambda x:x[0])
    kqua_=set(kqua)
    kqua__=list(kqua_)
    return kqua__

A=list(map(int,input().split()))
k=int(input())
print(cap(A,k))
```
